## Oppgave 3

```python
g = lambda x: 1-x**3
f = lambda x: 1/(1+x**2)
Nit = 40
n = np.arange(Nit)
xsol = [np.arange(Nit * 1.0), np.arange(Nit * 1.0)]

fig = plt.figure(figsize=(12,24))
fig.subplots_adjust(hspace=.5)
i=0
y = 0
for x in np.array([0.0, 1.0, 0.682, 1.153, 1.25, -1.25]):
    i=i+1
    y = x
    
    xsol[0][0] = x    
    xsol[1][0] = y

    ax = plt.subplot(12, 1, i)
    fig.add_subplot(ax)
    
    plt.title('$x_0 = ' + str(x) + '$')
    for j in range(0, Nit):
        
        xsol[0][j] = x
        x = g(x)
        xsol[1][j] = y
        y = f(y)
        
    ax.plot(n, xsol[0], label='$f(x) = 1-x^3$')
    ax.plot(n, xsol[1], label=r'$g(x) = \frac{1}{1+x^2}$')
    plt.ylabel('$x_n$')
    ax.axis([0, Nit-1, -0.5, 1.5])
    ax.grid()
    ax.set_xticklabels([])
    ax.legend()
```

## Oppgave 4

```python
f = lambda x: x**2+np.e**x-1;
df = lambda x: 2*x+np.e**x;

N = 8
for x in np.array([-3.0, -1.0, -0.1, 0.0, 0.1, 1.0, 3.0]):
    x_0 = x
    for n in range(N):
        x = x - f(x)/df(x)

    display(Math('$x_0 = ' + str(x_0) + '\implies x = ' + '%.25f' % x + '$'))
```

```python
f = lambda x: x**4-2*x**3-98*x**2+200*x-100;
df = lambda x: 4*x**3-6*x**2-196*x+200;

N = 8
for x in np.array([-9.0, -3.0, 3.0, 9.0]):
    x_0 = x
    for n in range(N):
        x = x - f(x)/df(x)

    display(Math('$x_0 = ' + str(x_0) + '\implies x = ' + '%.25f' % x + '$'))
```
